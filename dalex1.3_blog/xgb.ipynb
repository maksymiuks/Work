```python
import numpy as np
import pandas as pd
import xgboost as xgb
import pickle
```

```python
data = pd.read_csv("HRonehot.csv")
X = data[['gender', 'age', 'hours', 'evaluation', 'salary']]
y = data[['status']]
```

```python
y
```

```python
model = xgb.sklearn.XGBClassifier()


model = model.fit(X, y)
model.predict_proba(X)
```

```python
pickle.dump(model, open("xgboost_multiclass.pkl", "wb"), protocol=2)
```
